```
# Importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'Iris.csv' is in the current directory
df = pd.read_csv('Iris.csv')

# Calculate IQR for SepalWidthCm
Q1 = np.percentile(df['SepalWidthCm'], 25, interpolation='midpoint')
Q3 = np.percentile(df['SepalWidthCm'], 75, interpolation='midpoint')
IQR = Q3 - Q1

# Upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Removing outliers
df = df[(df['SepalWidthCm'] > lower_bound) & (df['SepalWidthCm'] < upper_bound)]

# Display the old and new shape
print("Old Shape:", df.shape)
print("New Shape:", df.shape)

# Boxplot to visualize the distribution
sns.boxplot(x='SepalWidthCm', data=df)
plt.show()  # Display the plot
```
